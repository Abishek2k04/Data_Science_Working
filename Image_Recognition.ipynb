```python
from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.color import rgb2gray
```

```python
vijay =os.listdir("imgrecz/vijay")
```

```python
sk=os.listdir("imgrecz/sk")
```

```python
adhi=os.listdir("imgrecz/adhi")
```

```python
limit=10
vijay_images=[None]*limit
j=0
for i in vijay:
    if (j<limit):
         vijay_images[j]=imread("imgrecz/vijay/"+i)
         j+=1
    else:
        break
```

```python
limit=10
adhi_images=[None]*limit
j=0
for i in adhi:
    if (j<limit):
         adhi_images[j]=imread("imgrecz/adhi/"+i)
         j+=1
    else:
        break
```

```python
limit=10
sk_images=[None]*limit
j=0
for i in sk:
    if (j<limit):
         sk_images[j]=imread('imgrecz/sk/'+i)
         j+=1
    else:
        break
```

```python
view = imshow(vijay_images[0])
plt.show()
```

```python
view2=imshow(adhi_images[2])
plt.show()
```

```python
vijay_gray =[None]*limit
j = 0
for i in vijay:
    if j < limit: 
        vijay_gray[j]=rgb2gray(vijay_images[j])
        j+=1
    else:
        break
```

```python
sk_gray = [None] * limit
j = 0
for i in sk:
    if j < limit: 
        sk_gray[j]=rgb2gray(sk_images[j])
        j+=1
    else:
        break
```

```python
adhi_gray = [None] * limit
j = 0
for i in adhi:
    if j < limit: 
        adhi_gray[j]=rgb2gray(adhi_images[j])
        j+=1
    else:
        break
```

## Now Flatten Image from (512,512) matrix to 262144,1 vector


```python
for i in range(len_of_images_vijay):
    vijay_gray[i]=np.ndarray.flatten(vijay_gray[i]).reshape(flatten_size_vijay,1)
```

```python
for i in range(len_of_images_sk):
    sk_gray[i]=np.ndarray.flatten(sk_gray[i]).reshape(flatten_size_sk,1)
```

```python
for i in range(len_of_images_adhi):
    adhi_gray[i]=np.ndarray.flatten(adhi_gray[i]).reshape(flatten_size_adhi,1)
```
